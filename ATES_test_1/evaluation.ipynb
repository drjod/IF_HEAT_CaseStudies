```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


SMALL_SIZE = 26
MEDIUM_SIZE = 30
BIGGER_SIZE = 34

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title

```

```python
output = pd.read_table("output_timeseries.csv", header=0, sep=",")

output
```

```python
fig, axes = plt.subplots(3, 1, figsize=(20, 15))
plt.subplots_adjust(wspace=.5, hspace=.4)

ax = axes[0]

ax.plot(output.Q_actual/1e6, 'o', label='Q_actual')
ax.plot(output.Q_target/1e6, 'o', label='Q_target')

ax.set_xlabel('[d]')
ax.set_ylabel('[MW]')

ax.grid(True)
ax.legend()

#####

ax = axes[1]

ax.plot(output.T_rf_sto, 'o', label='Return')
ax.plot(output.T_ff_sto, 'o', label='Feed')


ax.set_xlabel('[d]')
ax.set_ylabel('[Â°C]')

ax.grid(True)
ax.legend()

#####

ax = axes[2]

ax.plot(output.v_sto, 'o')


ax.set_xlabel('[d]')
ax.set_ylabel('[m/s?]')

ax.grid(True)
ax.legend()

#fig.savefig('results.png', dpi=300)
```
